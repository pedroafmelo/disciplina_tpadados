# ESCOLAS TPAD

## Nesta atividade, descobriremos o número de escolas da cidade de João Pessoa, após juntar os DataFrames disponíveis

```python
# Lendo os DataFrames com o pandas:

import pandas as pd

#DF1:

df1 = pd.read_csv("../dados/EDUCAÇÃO/fato_rendimento_escolar.csv", sep = "|", encoding = "latin", skiprows = 5).drop_duplicates()
df1.columns =["Ano", "Código da Escola", "TAxa de Abandono no Ensino Médio (%)"]

#DF2:

df2 = pd.read_excel("../dados/EDUCAÇÃO/dim_escola.xlsx", sheet_name= "dados", header= 0).drop_duplicates()

# Aqui, mudei alguns dados com o método map, para facilitar a leitura e análise:

dicio1 = {1: "Federal", 2: "Estadual", 3: "Municipal"}
dicio2 = {1: "Urbana", 2: "Rural"}
dicio3 = {1: "Em Atividade", 2: "Paralisada", 3: "Extinta (Ano do censo)", 4: "Extinta em anos anteriores"}

df2["cod_rede"] = df2["cod_rede"].map(dicio1)
df2["localizacao"] = df2["localizacao"].map(dicio2)
df2["situacao_funcionamento"] = df2["situacao_funcionamento"].map(dicio3)

#DF3: 

df3 = pd.read_csv("../dados/EDUCAÇÃO/dtb_municipios.csv", sep = ";", encoding = "latin").drop_duplicates()
```

```python
# Realizando a junção dos DataFrames com o método Merge, utilizando o inner com how.

df_new = (
    df1.merge(df2, how = "inner", left_on = "Código da Escola", right_on = "cod_escola", indicator = "merge_df2")
          .merge(df3, how = "inner", left_on = "cod_mundv", right_on = "cod_mundv" , indicator = "merge_df3")
          )

df_new
```

```python
# Realizando a query, para analisar apenas os municípios com nome João Pessoa 

data = df_new.query("nome_mun == 'João Pessoa'")

# Com o método unique, eliminamos as duplicatas, criando uma array, e
# com o size, descobrimos o tamanho da array criada.
data["nome_escola"].unique().size
```
